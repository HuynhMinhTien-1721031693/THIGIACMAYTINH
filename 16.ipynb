```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
#Đọc ảnh
image_path = 'C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/conmeo2.jpg'
image = Image.open(image_path)
#Ảnh xám
gray_image = image.convert('L')
#Chuyển ảnh xám về mảng
image_a = np.array(gray_image)
#Làm mờ ảnh xám với bộ lọc Gaussian
blurred_image_a = ndimage.gaussian_filter(image_a, sigma = 2, mode = "constant")
blurred_image_b = ndimage.gaussian_filter(image_a, sigma = 4, mode = "constant")
blurred_image_c = ndimage.gaussian_filter(image_a, sigma = 6, mode = "constant")
#Hiển thị ảnh
fig, axs = plt.subplots(2,2, figsize = (10, 8))
plt.subplot(2,2,1)
plt.imshow(gray_image, cmap = 'gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,2,2)
plt.imshow(gray_image_a, cmap = 'gray')
plt.title('sigma=2')

plt.subplot(2,2,3)
plt.imshow(blurred_image_b, cmap = 'gray')
plt.title('sigma=4')

plt.subplot(2,2,4)
plt.imshow(blurred_image_c, cmap = 'gray')
plt.title('sigma=6')
```
