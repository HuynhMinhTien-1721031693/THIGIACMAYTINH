```python
from scipy import ndimage
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
#c nh
image = Image.open('C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/Demconvat1.jpg')
#Chuyn sang nh xm
gray_image = image.convert('L')
image_array = np.array(gray_image) 
#Chuyn nh sang dng nh phn
binary_image = 1*(image_array<200)
#Gn nhn
labeled_image, num_feautures = ndimage.label(binary_image)
#To li subplot vi 3 dng 1 ct
fig, axs = plt.subplots(3,1, figsize = (50,50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap = 'gray')
plt.title('nh xm')

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap = 'gray')
plt.title('nh nh phn')

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap = 'gray')
plt.title(f" nh nhn ({num_feautures} i tng)")

plt.show();
```

```python
from PIL import Image
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt
#c nh
image = Image.open('C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/Demconvat1.jpg')
#Chuyn sang nh xm
gray_image = image.convert('L')
#Chuyn nh sang nh phn
binary_image = 1*(image_array<200)
#Thc hin opening  loi b nhiu v ni cc i tng
opened_image = ndimage.binary_opening(binary_image, structure = np.ones((5,5)))
#nh nhn
labeled_image, num_feautures = ndimage.label(opened_image)
#To li subplot vi 3 dng 1 ct
fig, axs = plt.subplots(4,1, figsize = (50,50))

plt.subplot(4,1,1)
plt.imshow(binary_image, cmap = 'gray' )
plt.title('nh xm')

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap = 'gray' )
plt.title('nh nh phn')

plt.subplot(4,1,3)
plt.imshow(opened_image, cmap = 'gray' )
plt.title('nh sau opening')

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap = 'gray' )
plt.title(f" nh nhn ({num_feautures} i tng)")

plt.show();
```
