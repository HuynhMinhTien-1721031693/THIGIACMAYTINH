```python
from scipy import ndimage
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
#Đọc ảnh
image = Image.open('C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/Demconvat1.jpg')
#Chuyển sang ảnh xám
gray_image = image.convert('L')
image_array = np.array(gray_image) 
#Chuyển ảnh sang dạng nhị phân
binary_image = 1*(image_array<200)
#Gắn nhãn
labeled_image, num_feautures = ndimage.label(binary_image)
#Tạo lưới subplot với 3 dòng 1 cột
fig, axs = plt.subplots(3,1, figsize = (50,50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap = 'gray')
plt.title('Ảnh xám')

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap = 'gray')
plt.title('Ảnh nhị phân')

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap = 'gray')
plt.title(f"Đã đánh nhãn ({num_feautures} đối tượng)")

plt.show();
```

```python
from PIL import Image
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt
#Đọc ảnh
image = Image.open('C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/Demconvat1.jpg')
#Chuyển sang ảnh xám
gray_image = image.convert('L')
#Chuyển ảnh sang nhị phân
binary_image = 1*(image_array<200)
#Thực hiện opening để loại bỏ nhiễu và nối các đổi tượng
opened_image = ndimage.binary_opening(binary_image, structure = np.ones((5,5)))
#Đánh nhãn
labeled_image, num_feautures = ndimage.label(opened_image)
#Tạo lưới subplot với 3 dòng 1 cột
fig, axs = plt.subplots(4,1, figsize = (50,50))

plt.subplot(4,1,1)
plt.imshow(binary_image, cmap = 'gray' )
plt.title('Ảnh xám')

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap = 'gray' )
plt.title('Ảnh nhị phân')

plt.subplot(4,1,3)
plt.imshow(opened_image, cmap = 'gray' )
plt.title('Ảnh sau opening')

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap = 'gray' )
plt.title(f"Đã đánh nhãn ({num_feautures} Đối tượng)")

plt.show();
```
